```python
# -> combine multiple models
# -> to improve overall performance of ML models
# Goal -> enhance predictive accuracy

# Weak Learner WL -> Base Learner
# -> is learning algorithm capable of producing classifiers with probability of error
# -> stictly less than random guessing (binary classifiction -> 0.5)
# -> performs lesser than random guessing

# Strong Learner SL 
# -> yield classifiers with small error probaility
# -> perform better than random guessing

# Categories in Ensemble Learning
# 1. Sequential Ensemble Technique
# 2. Paralled Ensemble Technique

# 1. Sequential Ensemble Technique
# -> train models sequentially with each model attempting to correct its predecessor score
# -> Model1 -> Model2 (correct errors of Model1) -> Model3 -> 
# -> refining predictions sequentially
# -> Boosting - e.g. 

# The steps involved in the sequential ensemble technique depend on the data structure and the requirements of the application. They are:
# Base Model Selection: Choose the initial model.
# Sequential Training: Train models one after the other.
# Error Correction: Each model learns from previous mistakes.
# Prediction Refinement: Iteratively refine predictions.
# Combining Predictions: Combine predictions from all models.
# Evaluation: Assess ensemble performance using metrics.

# 2. Paralled Ensemble Technique
# The steps involved in the parallel ensemble technique are:
# 1.Data Partitioning: Divide the dataset into subsets
# 2.Model Training: Train models concurrently on subsets
# 3.Prediction: Models make independent predictions
# 4.Combining Predictions: Aggregate predictions using techniques like voting or averaging
# 5.Evaluation: Assess ensemble performance using metrics

# Voting technique
# -> involves the aggreagting the predictions from multiple models to arrive at a final prediction
# 1. Majority Voting (Hard Voting)
# -> every model makes its own prediction
# -> whichever prediction has a majority vote -> that is considered as final prediction

# 2. Weightes Voting (Soft Voting)
# -> takes the estimates from each model
# -> every model will give the probaility of each class - model 1 - class 0 class1 class2
# -> average out all these probabilities
# -> final prediction : class with highest average probability is considered

```
