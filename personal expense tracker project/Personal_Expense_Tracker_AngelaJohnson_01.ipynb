# Angela Johnson's Personal Expense Tracker Project 

```python
# list will store all expenses as dictionaries
expenses = []

def add_expense():
    date = input("Enter the date (YYYY-MM-DD): ")
    category = input("Enter the category (e.g., Food, Travel): ")
    try:
        amount = float(input("Enter the amount spent: "))
    except ValueError:
        print("Invalid amount! Please enter a number.")
        return
    description = input("Enter a brief description: ")

    expense = {
        'date': date,
        'category': category,
        'amount': amount,
        'description': description
    }

    expenses.append(expense)
    print("Expense added successfully!")
    print(expense)  # Show what was added
```

```python
# call funtion 
add_expense()
```

```python
print(expenses)
```

```python
def view_expenses():
    if not expenses:
        print("No expenses recorded.\n")
        return

    print("Your recorded expenses:\n")
    for i, expense in enumerate(expenses, 1):
        print(f"{i}. On {expense['date']}, you spent ${expense['amount']} on {expense['category']} — {expense['description']}.")
```

```python
view_expenses()
```

```python
# monthly budget 

monthly_budget = 0.0 # storing the budget in this variable 

# function to set the monthly budget
def set_budget():
    global monthly_budget

    print("Set your monthly budget below.")
    user_input = input("Enter the total amount you want to budget for the month: ")

    try:
        monthly_budget = float(user_input)
        print("Your budget has been set to $" + str(monthly_budget))
    except:
        print("Uh oh that is invlaid. Please enter a valid NUMBER.")
```

```python
# tracking budget 

# function to track expenses and compare with budget
def track_budget():
    print("Tracking your budget... \n")

    # add up all the expenses
    total_spent = 0
    for expense in expenses:
        total_spent = total_spent + expense['amount']

    print("You have spent $" + str(total_spent) + " so far.")

    # compare with budget
    if total_spent > monthly_budget:
        print("\nYou have exceeded your budget!!!")
    else:
        remaining = monthly_budget - total_spent
        print("\nYou have $" + str(remaining) + " left for the month.")
```

```python
set_budget()
```

```python
track_budget()
```

```python
import csv 

# saving and loading expenses 
def save_expenses():
    file = open("expenses.csv", "w", newline="")
    writer = csv.writer(file)
    writer.writerow(["date", "category", "amount", "description"])  # header

    for expense in expenses:
        writer.writerow([expense["date"], expense["category"], expense["amount"], expense["description"]])  # each row

    file.close()
    print("Expenses have been saved to a file called 'expenses.csv'.")
```

```python
# load expenses from CSV file 

def load_expenses():
    try:
        file = open("expenses.csv", "r")
        reader = csv.DictReader(file)

        for row in reader:
            row["amount"] = float(row["amount"])  # amount was saved as text --> we convert it back
            expenses.append(row)

        file.close()
        print("Expenses loaded successfully from 'expenses.csv'.")

    except FileNotFoundError:
        print("No saved expenses file found yet.")
```

```python
# show menu funtion 
def show_menu():
    print("\n--- Personal Expense Tracker ---\n")
    print("1. Add expense\n")
    print("2. View expenses\n")
    print("3. Track budget\n")
    print("4. Save expenses\n")
    print("5. Exit\n")


# main funtion to excecute program 
def main():
    load_expenses()  # Load saved data at the beginning

    while True:
        show_menu()  # Show the menu options
        choice = input("Choose an option (1–5): ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            track_budget()
        elif choice == "4":
            save_expenses()
        elif choice == "5":
            save_expenses()
            print("Your expenses have been saved.")
            break
        else:
            print("Invalid option. Please choose a number from 1 to 5.")
```

```python
main()
```
