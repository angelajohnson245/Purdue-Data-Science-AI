```python
# loading and inspecting the dataset 

import pandas as pd

df = pd.read_csv("Smart watch prices.csv")

df.head()
df.shape
df.info()
df.isna().sum()
df.duplicated().sum()
```

### checking for 
* dataset size and structure
* column data types 
* missing values and duplicates 
* formating issues 


```python
df.columns
```

```python
# standardize the column names 
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
)

df.columns
```

```python
# clean and transform pricing data 
df['price_usd'] = (
    df['price_usd']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)
```

```python
# quick sanity check 
df[['price_usd']].describe()
```

> Column names were standardized to lowercase snake_case format to keep consistency and reduce errors during data processing. Pricing value swere cleaned and converted from string-based formats into numerical values to support statistical analysis and modeling.
