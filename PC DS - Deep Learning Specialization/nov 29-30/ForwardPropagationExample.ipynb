Import libraries
Get some random data
Predict the output using some equation of type y=mx+b
Calculate errors
Calculate the sumof errors

```python
import numpy as np
```

```python
x = np.random.randint(-1, 5, 11) # generate 11 nos from -1 to 4
y_pred = 3*x + 2
print(y_pred) # got the predicted values using y=3x+2
```

```python
# target value = 4
y = 4
error = abs(y_pred - y)
print(error)
```

```python
error_sq =abs(y_pred - y)**2
print(error_sq)
```

```python
print(error.sum()/11)
print(error_sq.sum()/11)
```
