Define a class called Neural Network
Random Generator 
Assign random weights 
tanh function to describe the S shaped curve

```python
class NeuralNetwork :
    def __init__(self) :
        np.random.seed(2) # reproducalibilty
        self.synaptic_weights = 2 * np.random.random((3,1))-1 
        # n.w has 3 inputs features -> 1 output
        # initialze weights randomly in the range [-1,1] of shape (3,1)
    def __tanh(self, x) :
        return np.tanh(x)
        # Activation function - tanh - squash vlues in (-1, 1) range# Normalize the output

    def __tanh_derivative(self, x):
        t = (np.exp(x) - np.exp(-x)) /(np.exp(x) + np.exp(-x))  # compute tanh manually
        return 1 - t**2 # tanh'x  = 1 - tanh^2(x)

    def think(self, inputs) :
        return self.__tanh(np.dot(inputs, self.synaptic_weights))

    def __train__(self, training_set_inputs, training_set_outputs, number_of_training_iterations) :
        for interation in range(number_of_training_iterations) :
            output = self.think(training_set_inputs) # forward pass
            error = training_set_outpus - output
            adjustment = np.dot(training_set_inputs.T, error*self.__tanh_derivative(output))
            self.synaptic_weights += adjustment
    
```
