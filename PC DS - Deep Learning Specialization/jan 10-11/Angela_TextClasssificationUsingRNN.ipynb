# Text Classification (NLP)

## What is Text Classification?
Text classification is the task of assigning **predefined labels** to text data.

### Examples
- **Email classification:** spam / not spam  
- **Sentiment analysis:** positive / negative review  
- **Topic classification:** news, sports, politics  

---

## Deep Learning for Text Classification
Deep learning enables machines to automatically learn patterns from text and perform classification tasks.

---

## Text Classification Pipeline

### Steps

### 1. Text Preprocessing
Preparing raw text before feeding it into a model.

- Lowercasing  
- Removing punctuation  
- Tokenization (breaking text into words)

Example:
"I liked this movie!"
→ ["i", "liked", "this", "movie"]

---

### 2. Text Representation (Embeddings)
Convert words into numerical vectors that capture semantic meaning.

Example:
"king" → [0.25, -0.61, 0.98, ...]


---

### 3. Deep Learning Model
Models learn patterns from embeddings.

- RNN (Recurrent Neural Network)
- Transformers (BERT)

➡ The model extracts meaningful patterns from embeddings.

---

### 4. Classification Layer (Output Layer)

- **Binary Classification:** Sigmoid activation  
  - Example: spam vs not spam

- **Multi-class Classification:** Softmax activation  
  - Example: positive / neutral / negative

---

### 5. Training Process
- Model learns by minimizing loss
- Weights updated using backpropagation
- Accuracy improves over multiple epochs

---

## RNN — Recurrent Neural Network

### What is an RNN?
An RNN processes **sequential data** where **order and context matter**.

---

### Key Characteristics
- Processes input one element at a time
- Maintains a **hidden state** (memory)
- Remembers information from previous steps
- Suitable for text and sequence data

---

### How RNN Works with Text
- Text → sentences → sequential data
- Reads text word by word
- Uses hidden state to remember past words

Example:
"I liked this movie" → Positive


---

### RNN Cell
- Takes current input
- Takes previous hidden state
- Produces a new hidden state

➡ Hidden state stores past information.

---

### RNN Input Flow
- Input text
- Tokenization
- Embeddings
- Sequential processing
- Hidden state plays a key role

---

### Why Use RNN for Text Processing?
- **Sequential Memory:** captures word order  
- **Hidden State:** stores context  
- **Weight Sharing:** learns general patterns efficiently

```python
# import libraries 
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = [
    'I liked this movie',
    'This fil was terrible',
    'Amazing experience',
    'Worst movie ever',
    'I enjoyed the movie',
    'I hate this movie'
]

labels = [1, 0, 1, 0, 1, 0]
```

```python
vocab_size = 1000 # limiting model number of words to 1000
max_length = 7  # every sentence will be converted into exactly 6 numbers 


tokenizer = Tokenizer(num_words = vocab_size, oov_token='<OOV>' )

# -> oov_token='<OOV>' -> handles unknown words 
# -> "blockchain is future " -> blockchain is a new word, the model will replace it iwth oov for it not to crash 

tokenizer.fit_on_texts(sentences) # read all sentences, build a word index
                                  # lower index 1 = more frequent word 
sequences = tokenizer.texts_to_sequences(sentences) # convert each word into numeric index 
                                                    # 'I liked the movie' -> [2,3,4,5]

# padding and truncation 
# shorter and longer sentences, shorter needs to be padded and linger needs to be truncated 

padded_sequences = pad_sequences(
    sequences,
    maxlen = max_length,
    padding = 'pre',
    truncating='pre'
)

# padding -> add 0s at the beginning if sentence is shorter than 6 
# truncating -> cut words from beginning if sentence is longer than 6 
```

```python
# X, y
X = np.array(padded_sequences, dtype='int32')  # input features: padded word index sequences
y = np.array(labels, dtype='int32')             # output labels (e.g., 0 = negative, 1 = positive)
```

```python
# stacking one after the other
model = Sequential([
    Embedding(
        input_dim=vocab_size,   # total number of unique words in the vocabulary
        output_dim=16            # each word becomes a 16-dimensional vector
    ),

    SimpleRNN(
        32                       # reads sentence word by word,
                                 # maintains a hidden state,
                                 # outputs a single vector of size 32
    ),

    Dense(
        1,                       # output layer with 1 neuron (binary classification)
        activation='sigmoid'     # sigmoid outputs probability between 0 and 1
    )
])
```

```python
# compile the model 
model.compile(
    optimizer = 'adam', 
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
# summmary 
model.summary()
```

# it is 0 bc we have not done training for it yet

```python
# fit the model (using x, y)
# run for 20

model.fit(
    X,
    y,
    epochs = 20,
    batch_size = 2, 
    verbose = 1
)
```

* accuracy is increasing and loss is decreasing

```python
test_sentences = ["I really loved this movie"]

test_seq = tokenizer.texts_to_sequences(test_sentences)
test_pad = pad_sequences(test_seq, maxlen=max_length)

prediction = model.predict(test_pad)

print("Prediction Score:", prediction)
print("Sentiment:", "Positive" if prediction[0][0] > 0.5 else "Negative")
```
