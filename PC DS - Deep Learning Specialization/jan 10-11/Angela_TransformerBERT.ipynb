# Transformer Models for NLP (Natural Language Processing)

## What is a Transformer?
A **Transformer** is a neural network architecture designed to process **sequential data (text)** using **attention mechanisms** instead of recurrence (RNNs) or convolution (CNNs).

---

## Core Idea
➡ Every word can **directly attend to every other word** in the sentence.

### Unlike RNNs (left → right reading), Transformers:
- Look at the **entire sentence at once**
- Learn **relationships between words** using attention
- Process text **in parallel**

---

## Why Attention Matters
- Captures long-range dependencies
- Understands context better than RNNs
- Faster training due to parallel processing

---

## Architecture of a Transformer Model

### Two Main Parts:

#### 1. Encoder
- Used for **understanding text**
- Produces contextual representations of words
- **Example:** BERT

#### 2. Decoder
- Used for **text generation**
- Predicts the next word in a sequence
- **Example:** GPT

---

## Summary
- Transformers use **attention instead of recurrence**
- Words interact with all other words in a sentence
- Encoder → understanding
- Decoder → generation


## Input Text Example (Transformer Flow)

### Input some text:
➡ **"I like Deep Learning"**

### Processing Steps:
➡ Tokenisation  
➡ Embeddings  
➡ Self-Attention  

---

### Why Self-Attention is Important
Self-attention helps the model understand **which words relate to each other** in a sentence.

### Example:
➡ *"The animal did not cross the street because it was tired."*

**Question:**  
➡ What does **"it"** refer to?

**Answer:**  
➡ **"animal"**

✔ The transformer uses **attention** to correctly link **"it" → "animal"** by analyzing context across the entire sentence.

---

### Key Idea
- Transformers look at **all words at the same time**
- Attention determines **word relationships**
- Context is captured more effectively than sequential models

# BERT (Bidirectional Encoder Representations from Transformers)

## What is BERT?
**BERT** is a **Transformer-based, encoder-only model** designed to understand language by capturing **context from both directions**.

---

## Key Features of BERT
➡ Transformer-based **encoder-only** architecture  
➡ Understands **language context**  
➡ Looks at both **left and right sides** of a word simultaneously  
➡ Reads text **bidirectionally**  
➡ Understands **true contextual meaning**

---

## Why Bidirectional Context Matters

### Example:
➡ *"He went to the bank to deposit money."*  
➡ *"He sat on the bank of a river."*

✔ BERT understands that the word **"bank"** has **different meanings** in each sentence based on surrounding context.

---

## How BERT Achieves This
➡ Uses a **Transformer Encoder**  
➡ Applies **multi-layer, bidirectional self-attention**  
➡ Every word attends to **all other words** in the sentence

---

## Summary
- BERT captures **true context** using bidirectional attention
- Same word can have **different meanings** depending on usage
- Ideal for **text understanding tasks** (classification, QA, NER)

```python
# Text Classification using BERT Model

import pandas as pd              # data manipulation (tables, CSV files, DataFrames)
import torch                     # core PyTorch library for tensors and deep learning
import numpy as np               # numerical computations and array operations

from transformers import BertTokenizer, BertModel
# BertTokenizer -> converts text into tokens (IDs) that BERT understands
# BertModel -> pre-trained BERT model (outputs embeddings / hidden states)

from torch import nn
# nn -> neural network module (used to define layers like Linear, Dropout, Loss functions)

from torch.optim import Adam
# Adam -> optimizer that updates model weights during training

from tqdm import tqdm
# tqdm -> progress bar for loops (commonly used during training epochs)
```

```python
df = pd.read_csv('bbc-text.csv')
df.head()
```

```python
# Encode labels
labels = df['category'].astype('category').cat.codes
labels_dict = dict(enumerate(df['category'].astype('category').cat.categories))
print(labels_dict)
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    df['text'],
    labels, 
    test_size=0.2,
    random_state=42
)

tokenize = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize(texts):
    return tokenizer(
        list(texts),
        padding = True, 
        truncation = True, 
        max_length = 128, 
        return_tensors = 'pt'
    )
    

train_encodings = tokenize(X_train)
val_encodings = tokenize(X_val)
```

* Tokenizes raw text into numerical tensors

```python
# Create PyTorch Dataset

class BBCDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels.reset_index(drop=True)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)


# create dataset objects
train_dataset = BBCDataset(train_encodings, y_train)
val_dataset = BBCDataset(val_encodings, y_val)
```

* Wraps those tensors and labels into a PyTorch Dataset so they can be fed into a DataLoader for training;

**How you know which one to use:**

* Use tokenization when you’re converting text → numbers (always comes first)

* Use Dataset when you’re ready to train or validate a model (always comes after)

_Think of it like this_

[9] “Prepare the data”

[10] “Package the data for training”

```python
# create Bert classifier 

class BertClassifier(nn.Module) :
    def __init__(self, num_classes):
        super(BertClassifier, self).__init__()
        self.bert = BertModel.from_pretrained("bert-base-uncased")
        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(768, num_classes)

    def forward(self, input_ids, attention_mask) :
        outputs = self.bert(input_ids = input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        x = self.dropout(pooled_output)
        return self.fc(x)
```

```python
# Training Setup

device = torch.device("cpu")
model = BertClassifier(num_classes = len(labels_dict))
model.to(device)
optimizer = Adam(model.parameters(), lr=2e-5)
loss_fn = nn.CrossEntropyLoss()
```

```python
# training loop 

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=8)

epochs = 3

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for batch in tqdm(train_loader):
        optimizer.zero_grad()

        input_ids = batch["input_ids"].to(device).long()
        attention_mask = batch["attention_mask"].to(device).long()
        labels = batch["labels"].to(device).long()

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        loss = loss_fn(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    
    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch + 1}/{epochs}, Training Loss: {avg_loss:.4f}")
```

```python
# Validation Accuracy

model.eval()
correct = 0
total = 0

with torch.no_grad():
    for batch in val_loader:
        input_ids = batch["input_ids"].to(device).long()
        attention_mask = batch["attention_mask"].to(device).long()
        labels = batch["labels"].to(device).long()

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        pred = torch.argmax(outputs, dim=1)
        correct += (pred == labels).sum().item()
        total += labels.size(0)

accuracy = correct / total
print(f"Validation Accuracy: {accuracy:.4f}")
```

```python
# Predict on New Text 

def predict(text):
    enc = tokenizer(
        text,
        return_tensors = 'pt', # when using torch-> it should return pytorch tensors 
        truncation = True, 
        padding = True, 
        max_length = 128
    )
    
    enc = {k:v.to(device) for k,v in enc.items()} # k: key, dicitionary key | v: value corresponding tensor value
    
    with torch.no_grad():
        enc.pop('token_type_ids', None)
        outputs = model(**enc) # ** -> entire dictionary 
        pred = torch.argmax(outputs, dim = 1).item()
    

    return labels_dict[pred]


print(predict('The stock market showed good growth today'))
    
```
