How to implement ensemble techniques using Keras and sklearn

```python
import sklearn
from sklearn.datasets import make_moons # generate some dataset - samples no....
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X,y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
log_clf = LogisticRegression(solver='lbfgs', random_state=42)
# solver='lbfgs' - optimization algo -> find weights that minize loss
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
svm_clf = SVC(gamma='scale', random_state=42, probability=True)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

def build_nn():
    model = Sequential([
        Dense(50, activation='relu', input_shape=[2]),
        Dense(1, activation='sigmoid')
    ])
    model.compile(
          optimizer="Adam",
          loss="binary_crossentropy",
          metrics=['accuracy'])
    return model
```

```python
# Wrap scikit learn and build nn with 500 epachs with Keras clasifier
!pip install scikeras
```

```python
from scikeras.wrappers import KerasClassifier
keras_clf = KerasClassifier(model = build_nn, epochs=50, verbose=0, random_state=42)
```

```python
# keras_clf._estimator_type = "classifier"
```

```python
# Create voting classifier

voting = VotingClassifier(
           estimators = [('lr',log_clf),
                         ('rf', rf_clf),
                         ('svc', svm_clf),
                         ('keras', keras_clf)],
           voting = 'soft',
           flatten_transform = True)
                          
```

```python
voting.fit(X_train, y_train)
```
