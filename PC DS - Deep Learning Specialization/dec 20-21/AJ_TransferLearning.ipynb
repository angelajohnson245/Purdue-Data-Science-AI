```python
# Transfer Learning
# Pre Trained models - reuse
# Image Classification, audio, video - many pre trained models

# Transfer Learning - technique in ML - pre trained model - trained on large dataset
# -> starting point for solving some related problems

# Import libraries
# Add classifier layers
# Preprocessing, feature extraction

from tensorflow.keras.utils import load_img  # Load an image from the file system
from tensorflow.keras.utils import img_to_array  # convert an image into Numpy Array
from keras.applications.vgg16 import preprocess_input  # VGG16 - extract input images to be preprocessed in a specific way
from keras.applications.vgg16 import VGG16  # model architecture, CNN model, pre-trained on ImageNet dataset, image classification 
```

```python
# Add classifier layers
# -> demonstrate how to load pre-trained VGG16 model wihtout its classifier layers and then add new custom elayers 
# -> new modesl is defined by connecting the output of pre trsined VGG16 tto a lfattneing layer 
# -> dense layer 
# -> multi class classification -> softmax 
```

```python
from keras.models import Model
from keras.layers import Dense
from keras.layers import Flatten 


model = VGG16(include_top = False, input_shape = (300,300,3))
# include_top = False -> remove the orginalclassifier, keep only convolution + pooling layers 
flat1 = Flatten()(model.layers[-1].output)
class1 = Dense(1024, activation='relu')(flat1)
output = Dense(10, activation='softmax')(class1)


model = Model(inputs = model.inputs, outputs=output)
model.summary()
```

* we added flatten, dense, dense_1

```python
# perform preprocessing and image extraction 
# -> load some image from disk and preprocess it to meet the input requrirements of VGG16 model 
# -> resizing,reshapning, img to array 
# -> modefifed model predict and extract featured from input image, result in a feature vector with some specific shape 
```

```python
import numpy as np

# dog 
image = load_img('dog.jpg', target_size = (300,300))
image = img_to_array(image)
image = np.expand_dims(image, axis=0)
image = preprocess_input(image)
predictions = model.predict(image)
predictions_class = np.argmax(predictions, axis=1)
print('Predicited class index: ', predictions_class)
print('Predicition probalilties: ', predictions)
```

```python
confidence = np.max(predictions)
print(confidence)
if confidence > 0.8 :
    print("High Confidence Prediction")
else :
    print("Low Confidence Prediction")
```
