# 🧠 Dropout — Regularization Technique in Neural Networks

## ⭐ What is Dropout?

* **Dropout** is a **regularization technique** used in neural networks.
* Its main purpose is to **prevent overfitting**.
* It works by **randomly disabling (dropping)** a subset of neurons during training.

---

## ❓ Why Do We Need Dropout?

### Deep Neural Networks:

* Have **millions of parameters**.
* Can easily **memorize** the training data instead of learning patterns.
* Neurons begin to **co-adapt**:

  * They rely too much on specific other neurons.
  * This creates fragile dependencies.
  * Result → **overfitting**.

### Dropout Fixes This:

* By randomly dropping neurons, dropout **breaks these dependencies**.
* Forces the network to learn **robust, distributed representations**.

---

## 🎯 Main Idea of Dropout

During **each training step**:

1. Randomly turn off (drop) some neurons.
2. Each iteration produces a **slightly different sub-network**.
3. This is like training **many smaller networks** and then **averaging** them at the end.

---

## 🔧 How Dropout Works (Conceptually)

For **every neuron** in the layer:

1. Generate a random number `r` in ([0,1]).
2. If `r < dropout_probability` → neuron is **dropped**

   * Output becomes **0**.
3. Else → neuron continues the normal forward pass.
4. Only the **remaining neurons** participate in the computation.

---

## 🎛 Key Terms

### **Dropout Probability (p)**

* Probability of **dropping** a neuron.
* Example:

  * `dropout = 0.5` → drop **50%** of neurons.

### **Keep Probability (1 - p)**

* Probability of **keeping** a neuron active.
* Example:

  * `keep_prob = 0.5` when dropout = 0.5.

---

## 📌 Example

* If `dropout = 0.5`, then on average:

  * **Half of the neurons** are removed during each training step.
  * The remaining neurons must learn to perform well **without depending** on the ones that were dropped.

---

```python
import numpy as np

def dropout_forward(x, dropout_rate=0.5, training=True):
    if not training:
        return x

    keep_prob = 1 - dropout_rate
    mask = (np.random.rand(*x.shape) < keep_prob) # this line is a lil extra 
    return x * mask
```

```python
x = np.array([1.0, 2.0, 3.0, 4.0])
for _ in range(5):
    print(dropout_forward(x, dropout_rate=0.5, training=True))
```

```python
def dropout_forward(x, dropout_rate1=0.5, training=True):
    if not training:
        return x
    
    # continuous random numbers between 0 and 1
    mask = np.random.rand(*x.shape)
    return x * mask
```

```python
x = np.array([1.0, 2.0, 3.0, 4.0])

for _ in range(5):
    print(dropout_forward(x, dropout_rate1=0.5, training=True))
```
