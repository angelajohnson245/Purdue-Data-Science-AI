```python
# most widely used optimizers in DL models
# adam - adaptive graduent based opitmization algorithim 
# combines the best ideas of:
# -> momentum: smooth and fast convergence 
# -> RMSProp: adaptive learning rate per paramter 


# maintains two moving averages for each paramenter
# 1. First Moment (mean) - like momentum 
# 2. Second Moment (uncentered varaince) - like RMSProp
# -> applies bias correction 
# -> uses adaptive learning rates 
# -> update paramters smoothly and efficiently 

# equation for first moment: mt = b1*m(t-1) + (1-b1)gt        # gt = delta of L(theta) 
# equation for second moment:  RMSProp 
# -> vt = b2*v(t-1) + (1-b2)gt**2
# bias correction -> reliable early learning 


# moves fast on flat surfaces 
# slows down near minima 
# handles nosiy gradients also 

# most widely used optimizer 
```

```python
import numpy as np 

def objective(x):
    return x**2 
    
def gradient(x):
    return 2*x
```

```python
def adam(objective, gradient, initial_x, alpha=0.001, beta1=0.9, beta2=0.999, eps=1e08, n_iter=100):
    x = initial_x
    m = 0.0
    v = 0.0 
    t = 0 
    history = []
    for i in range(n_iter):
        t += 1
        g = gradient(x)
        # calculate first moment 
        m = beta1 * m + (1-beta1) * g 
        # calculate second moment 
        v = beta2 + v + (1-beta2) * (g**2)
        # bias correction 
        m_corr = m / (1 - beta1 ** t)
        v_corr = v / (1 - beta2 ** t)
        # update (of x)
        x = x - alpha * m_corr / (np.sqrt(v_corr) + eps)
        history.append(x)
        if i%10 == 0:
            print(f'Iter {i} | x = {x} | loss = {objective(x)}')
    return x, history 
            
```

```python
# run Adam 

final_x, history = adam(
    objective = objective,
    gradient = gradient,
    initial_x = 10.0,
    n_iter=100
)

print('Final x : ', final_x)
```
