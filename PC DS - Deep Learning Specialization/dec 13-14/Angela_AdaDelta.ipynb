```python
import numpy as np 

def objective(x):
    return x**2 
    
def gradient(x):
    return 2*x
```

```python
def adadelta(objective, gradient, intial_x, rho=0.95, eps=1e-6, n_iter=100):
    x = intial_x
    
    # running averages
    eg2 = 0.0      # E[g^2]
    edx2 = 0.0     # E[dx^2]
    
    history = []
    
    for i in range(n_iter):
        g = gradient(x)
        
        # accumulate gradient squared
        eg2 = rho * eg2 + (1 - rho) * (g ** 2)
        
        # compute update (Adadelta rule)
        dx = - np.sqrt(edx2 + eps) / np.sqrt(eg2 + eps) * g
        
        # update running avg of update squared
        edx2 = rho * edx2 + (1 - rho) * (dx ** 2)
        
        # update parameter
        x += dx
        history.append(x)
        
        if i % 10 == 0:
            print(f"Iter {i} | x = {x} | loss = {objective(x)}")
    
    return x, history

        
```

```python
# call it --> run optimization 

final_x, history = adadelta( 
                            objective=objective,
                            gradient=gradient,
                            intial_x=10.0
                            )

print('Final x: ', final_x)
```

# eg2 vs edx2 
| Variable   | Meaning                         | Used For                                                                                  |
| ---------- | ------------------------------- | ----------------------------------------------------------------------------------------- |
| **`eg2`**  | Running average of gradient²    | Measures slope steepness → controls *denominator* (shrinks steps when gradients are big)  |
| **`edx2`** | Running average of update step² | Measures past step size → controls *numerator* (expands steps when updates are too small) |
