```python
# imports
from numpy import asarray, arange
from numpy.random import rand, seed
from numpy import arange
from numpy import meshgrid
import numpy as np
from matplotlib import pyplot 
```

```python
def objective(x,y):
    return x**2.0 + y**2.0

# derivative funtion 
def derivative(x,y):
    return asarray([x*2.0, y*2.0])
```

```python
# grad descent
from math import sqrt

def rmsprop(objective, derivative, bounds, n_iter, step_size, rho):

    # generate initial point
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])

    # list of the sum square gradients for each variable
    sq_grad_avg = [0.0 for _ in range(bounds.shape[0])]

    # run the gradient descent
    for it in range(n_iter):

        # calculate gradient
        gradient = derivative(solution[0], solution[1])

        # update squared gradient accumulators
        for i in range(len(gradient)):
            sg = gradient[i] ** 2.0
            sq_grad_avg[ i] = (sq_grad_avg[i] * rho) + (sg * (1.0-rho))

        # build new solution
        new_solution = []
        for i in range(len(solution)):
            # Adagrad step size
            alpha = step_size / (1e-8 + sqrt(sq_grad_avg[i]))

            # update variable
            value = solution[i] - alpha * gradient[i]
            new_solution.append(value)

        # update solution vector
        solution = asarray(new_solution)

        # evaluate candidate point
        solution_eval = objective(solution[0], solution[1])

        # print progress
        print(f'{it+1}, {solution} : {solution_eval}')

    return solution, solution_eval


# ----------------------------------------
# RUNNING IT
# ----------------------------------------
seed(42)
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])
n_iter = 40
step_size = 0.1
rho = 0.9     # common default


best, score = rmsprop(objective, derivative, bounds, n_iter, step_size, rho)
print(f'\nBest solution: {best} : {score}')
```
