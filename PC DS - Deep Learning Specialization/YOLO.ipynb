```python
from ultralytics import YOLO
from IPython.display import Image, display

# Load pretrained model - YOLOv8 - small & fast
model = YOLO("yolov8n.pt") # nano model, trained on common objects - person, car, bus, dog - 80

# Run detection on an image URL
results = model("https://ultralytics.com/images/bus.jpg", save=True)

# Display output image
display(Image(filename=results[0].save_dir + "/bus.jpg")
```
